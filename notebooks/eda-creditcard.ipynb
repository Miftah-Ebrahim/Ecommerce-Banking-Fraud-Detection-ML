# Exploratory Data Analysis: Credit Card Fraud Dataset

This notebook analyzes the secondary dataset `creditcard.csv` to understand class imbalance and feature distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Set style
sns.set_style("whitegrid")
```

## 1. Load Data

```python
try:
    df = pd.read_csv("../data/raw/creditcard.csv")
    print(f"Data loaded successfully. Shape: {df.shape}")
    print(df.head())
except FileNotFoundError:
    print("Error: creditcard.csv not found in ../data/raw/")
```

## 2. Class Imbalance

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df)
plt.title('Class Imbalance (0: Legitimate, 1: Fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

print(df['Class'].value_counts(normalize=True))
```

## 3. Distribution of Amount and Time

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 5))

# Amount Distribution
sns.histplot(df['Amount'], bins=50, kde=True, ax=ax[0])
ax[0].set_title('Distribution of Transaction Amount')
ax[0].set_xlim([0, 2000])  # Limit x-axis to see the bulk of data

# Time Distribution
sns.histplot(df['Time'], bins=50, kde=True, ax=ax[1])
ax[1].set_title('Distribution of Transaction Time')

plt.show()
```

## 4. Analysis Summary
- The dataset is highly imbalanced.
- Amount distribution is right-skewed.
- Time feature shows cyclical patterns likely related to time of day.
